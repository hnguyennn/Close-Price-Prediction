# Random Forest model testing

```python
# Loading necessary packages
%pip install scikit-learn

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report, accuracy_score 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Loading math for square root operator
import math
```

# Splitting columns into features and targets

```python
# Checking the column names
numerical_data.columns
```

```python
# Splitting columns into features or targets
X = numerical_data.drop(columns=['ClosePrice'])
Y = numerical_data['ClosePrice']
```

# Splitting into Training and Testing 

```python
#80% Training, 20% Testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Check the number of rows in each set
print(f"Training set size: {X_train.shape[0]} rows")
print(f"Testing set size: {X_test.shape[0]} rows")
```

# Goal MAE and MSE

```python
# MAE is aiming for 10-20% of the mean
# Using RMSE, the root squared of mse, to easily interpret, it is aiming for 10-20% of the mean as well

mean_y_train = Y_train.mean()
mean_y_test = Y_test.mean()

print(f"Y_train's mean is: {mean_y_train}\n")
print(f"Y_test's mean is: {mean_y_test}\n")

goal_train = [mean_y_train*0.1, mean_y_train*0.2]
goal_test = [mean_y_test*0.1, mean_y_test*0.2]

print(f"The goal MAE/RMSE for TRAINED DATA is between {goal_train[0]} and {goal_train[1]}\n")
print(f"The goal MAE/RMSE for TEST DATA is between {goal_test[0]} and {goal_test[1]}\n")
```

# Training Random Forest Model

```python
# creating a function to easily evaluate the model for future uses

def evaluate_model(Y_pred_test0, Y_pred_train0, num):
    # Calculating different metrics

    print(f"Training Model {num} Tests\n")
    print(f"The goal MAE and RMSE for TRAINING DATA is between {goal_train[0]} and {goal_train[1]}")
    print(f"The goal MAE and RMSE for TESTING DATA is between {goal_test[0]} and {goal_test[1]}\n")

    # Calculate Mean Absolute Error (MAE)
    mae_test = mean_absolute_error(Y_test, Y_pred_test0)
    mae_train = mean_absolute_error(Y_train, Y_pred_train0)
    print(f"Mean Absolute Error (TEST DATA): {mae_test}")
    print(f"Mean Absolute Error (TRAINING DATA): {mae_train}\n")


    # Calculate Root Mean Squared Error (RMSE)
    mse_test = mean_squared_error(Y_test, Y_pred_test0)
    mse_train = mean_squared_error(Y_train, Y_pred_train0)
    print(f"Mean Squared Error (TEST DATA): {mse_test}")
    print(f"Mean Squared Error (TRAINING DATA): {mse_train}\n")

    rmse_test = math.sqrt(mse_test)
    rmse_train = math.sqrt(mse_train)
    print(f"Root Mean Squared Error (TEST DATA): {rmse_test}")
    print(f"Root Mean Squared Error (TRAINING DATA): {rmse_train}\n")

    # Calculate R-squared (goodness of fit)
    r2_test = r2_score(Y_test, Y_pred_test0)
    r2_train = r2_score(Y_train, Y_pred_train0)
    print(f"R-squared (TEST DATA): {r2_test}")
    print(f"R-squared (TRAINING DATA): {r2_train}")
```

```python
# Initializing model
rf_model = RandomForestRegressor(
    n_estimators=50,  # Set number of trees to 30
    max_depth=15,      # Limit tree depth to 15
    min_samples_split=10,  # Increase min samples for split
    min_samples_leaf=5,    # Increase min samples for leaf node
    max_features='sqrt',   # Use square root of features for splitting
    random_state=42,       # Ensure reproducibility
    n_jobs=-1              # Use all available cores
)
```

```python
# Fit model using training data
rf_model.fit(X_train, Y_train)
 
```

```python
# Making prediction based on model
# Inputting in both the trained data, as well as the test data
Y_pred_test = rf_model.predict(X_test)
Y_pred_train = rf_model.predict(X_train)
```

```python
# Evaluating the model
evaluate_model(Y_pred_test, Y_pred_train, 1)
```

The Mean Absolute Error seems reasonable, since its a low number relative to the goal range. Since the MAE for the training data is lower than the MAE for the testing data, it might be caused by overfitting, where the model is too trained on the patterns of the trained data, and doesn't generalize well with new data. 

However, the Root Mean Squared Error for both the test and training data are a lot higher than the Mean Absolute Error. I think this is caused by outliers in the data, since the MSE is sensitive to larger errors than the MAE.

So I think it is overfitting, because it learned the patterns of the training data too closely, so when given new large outliers, it will have a large error. So I will try to reduce overfitting by decreasing max_depth and increasing min_samples_leaf. I will also increase n_estimators to see if the overall performance will improve, and keep everything else the same.
